Tratar os Dados

```
#Vou buscar 5 ações para analise
tickers = ['ABEV3.SA', 'WEGE3.SA','PETR4.SA','VALE3.SA','^BVSP']

dados = web.get_data_yahoo(tickers, period='5y')['Adj Close'] # Definir o periodo de 5 anos
```

```
dados.head()
```

```
retorno = dados.pct_change()# calcular diario de todos os ativos
```

```
retorno
```

```
retorno_acumulado = (1 + retorno).cumprod()

#tirar o NAN, colocando valor 1 nos NAN

retorno_acumulado.iloc[0] = 1 # na primeira da linha


retorno_acumulado
```

```
carteira  = 10000 * retorno_acumulado.iloc[:, :5]
carteira

carteira['saldo'] = carteira.sum(axis=1) # Criando a primeira coluna
carteira['retorno'] = carteira['saldo'].pct_change() # Criando a segunda coluna

carteira
```

ANALISE DE DADOS

```
pf.create_full_tear_sheet(carteira['retorno'], benchmark_rets=retorno['PETR4.SA'])
```
