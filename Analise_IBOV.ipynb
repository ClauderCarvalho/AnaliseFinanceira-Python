Obter os dados do Mercado

```
ibov = web.get_data_yahoo('^BVSP', start='2018-05-03', end='2020-04-01')
```

Criando um grafico do preço de fechamento 

```
ibov['Close'].plot(figsize=(22,8))
plt.show()
```

```
# Vamos pegar um intervalo Menor 
# Criar uma variavel 

ibov_fatiado = ibov[(ibov.index.year >= 2008) & (ibov.index.year <=2018)]# criamos um filtro do ano de 2000



# Calcular a media de 21 dias 
ibov_fatiado['Close'].plot(figsize=(22,8), label='IBOV')
ibov_fatiado['Close'].rolling(21).mean().plot(label='MM21')
ibov_fatiado['Close'].rolling(200).mean().plot(label='MM200')
plt.legend()


plt.show()
```

Vamos pegar todos os dias do mês de dezembro de toda a nossa base de dados 

```


ibov_fatiado = ibov[ibov.index.month == 12]
ibov_fatiado.tail(50)  # Criando um Filtro 
```

Pegar agora todo o dia 14 da base de dados 

```


ibov_fatiado = ibov[ibov.index.day == 14]
ibov_fatiado.tail(50)

```
